```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# will resolve driver compatibility issues
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver = web_driver()
```

```python
titles = []
ratings = []
prices = []
total_rating_counts = []
image_urls = []
delivery_times = []
```

```python
url="https://www.amazon.com/s?k=area+rug&i=kitchen-intl-ship&crid=3D1S1PL1X2OS0&sprefix=ar%2Ckitchen-intl-ship%2C469&ref=nb_sb_ss_ts-doa-p_1_2"
for page in range(1, 11):
    driver.get(url.format(page))
    time.sleep(3)
    products = driver.find_elements(By.CLASS_NAME, 's-result-item')
    for product in products:
        try:
            delivery_time = product.find_element(By.CSS_SELECTOR, '.a-color-base').text
        except:
            delivery_time = 'N/A'
        delivery_times.append(delivery_time)
        try:
            title = product.find_element(By.CSS_SELECTOR, '.a-text-normal').text
        except:
            title = None
        titles.append(title)
        try:
            total_rating_count = product.find_element(By.CLASS_NAME, '.a-declarative').text
        except:
            total_rating_count = None
        total_rating_counts.append(total_rating_count)
        try:
            price = product.find_element(By.CSS_SELECTOR, '.a-price').text
        except:
            price = None
        prices.append(price)
        try:
            image_url = product.find_element(By.CSS_SELECTOR, '.s-image').get_attribute('src')
        except:
            image_url = None
        image_urls.append(image_url)
        try:
            rating = product.find_element(By.CSS_SELECTOR, '.a-size-base puis-bold-weight-text').text
        except:
            rating = None
        ratings.append(rating)
        time.sleep(2)
driver.quit()
df = pd.DataFrame({
    'Title': titles,
    'Price': prices,
    'Rating': ratings,
    'Total Rating Count': total_rating_counts,
    'Delivery Time': delivery_times,
    'Image URL': image_urls
})
df.to_csv('amazon_products.csv', index=False)
data = pd.read_csv('amazon_products.csv')
data
```
